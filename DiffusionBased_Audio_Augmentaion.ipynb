```python
prompts=enh_prompts+distor_prompts
tr_len = [0.05,0.06,0.07,0.08,0.09,0.1,0.12,0.14,0.15,0.16,0.17,0.18]
tr_len = [0.15,0.16,0.17,0.18,0.19,0.20]
ddims = [10,20,30,40,50]
ddims = [40,50,60,70,80]
```

```python
!mkdir output_test
def save_waveform_n(waveform,name):
    if len(waveform) == 1:
        waveform = waveform[0]
        
    path = 'output/'
    os.makedirs(path, exist_ok=True)
    file_name = path+'/'+name
    sf.write(file_name, waveform.T, 16000)
    
```

```python
ii = 0
for ad in tqdm(all_audios):
    if (ii%200==0):
        print(ii)
    prom = random.choice(prompts)
    transfer_strength = random.choice(tr_len)
    ddi = random.choice(ddims)
    waveform = audioldm.style_transfer(
        model_audioldm,
        prom,
        ad,
        transfer_strength,
        random.randint(0, 10_000_000),
        duration=10.0,
        guidance_scale=2.5,
        ddim_steps=ddi,
        batchsize=1
    )
    name= ad.split('/')[1]
    save_waveform_n(waveform,name)
    ii+=1
```
